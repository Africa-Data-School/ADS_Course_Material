```python
#import matplotlib.style as mplstyle
#mplstyle.use('fast')
import numpy as np
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

# Scatter Plots

```python

x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='something', color='k', s=500, marker="+")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Graph')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
days = [1,2,3,4,5]

walking = [7,8,6,11,7]
Training =   [2,3,4,3,2]
partying =  [7,8,7,2,2]
doing_nothing =  [8,5,7,8,13]


plt.plot([],[],color='m', label='walking', linewidth=5)
plt.plot([],[],color='c', label='Training', linewidth=5)
plt.plot([],[],color='r', label='partying', linewidth=5)
plt.plot([],[],color='k', label='doing_nothing', linewidth=5)

plt.stackplot(days, walking,Training,partying,doing_nothing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stacked Graph')
plt.legend()
plt.show()

#All we did here was plot some empty lines, giving them the same colors, 
#and the correct labels in accordance with our stack plot. We also gave them a linewidth of 5, 
#to make the lines a bit thicker in the legend.u
```

```python
slices = [7,2,2,13]
teams = ['Afc','Sofapaka','Tusker','Gormahia']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=teams,
        colors=cols,
        #startangle=90,
        #shadow= True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
#. Here we create a subplot of 2 rows by 2 columns and display 4 different plots in each subplot.
fig,a =  plt.subplots(2,2)

import numpy as np
x = np.arange(1,5)

a[0][0].plot(x,x*x)
a[0][0].set_title('square')

a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')

a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')

a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
#plt.subplot(subplot(nrows, ncols, index)
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.10, 0.1, 0.8, 0.8]) # main axes           r  phl   w  h
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```

```python
#Plt.subplot2grid(shape, location, rowspan, colspan)

import matplotlib.pyplot as plt
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)
import numpy as np
x = np.arange(1,10)
a2.plot(x, x*x)
a2.set_title('square')
a1.plot(x, np.exp(x))
a1.set_title('exp')
a3.plot(x, np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```
